```python
import xml.etree.ElementTree as ET
import glob
import os
import pandas as pd
```

```python
def get_trans_id(file):
    fname = os.path.basename(file)
    trans_id = fname.split('.')[0]
    return trans_id
```

```python
def get_topic_times(xml):
    tree = ET.parse(xml)
    root = tree.getroot()
    time_tuples = [(float(top.attrib['starttime']), float(top.attrib['endtime'])) for top in root.findall('.//top')]
    return time_tuples
```

```python
def get_topic_text(times, xml):
    start = times[0]
    end = times[1]
    # print(start)
    # print(end)
    #print(start.type())
    tree = ET.parse(xml)
    root = tree.getroot()
    matching_segments = []
    for segment in root.findall(".//Segment"):
        start_time = float(segment.get("StartTime"))
        end_time = float(segment.get("EndTime"))
        if start <= end_time and end > start_time:
            # segment_content = "".join(segment.itertext()).strip()
            # matching_segments.append(segment_content)
            matching_segments.append(segment.text.strip())
    #print(matching_segments)
    # Store the text within matching segments
    # topic_sents = [segment.text.strip() for segment in matching_segments]
    # return topic_sents
    return matching_segments
    #print(matching_segments[45].text)
```

```python
TRANS_PATH = 'raw_transcript_data/ICSI_original_transcripts/transcripts/'
TOPIC_PATH = 'raw_transcript_data/ICSIplus/Contributions/GalleyTopics/'
```

```python
topic_files = glob.glob(TOPIC_PATH + '/*.xml')
# topic_fnames = [os.path.basename(f) for f in topic_files]
# trans_ids = [s.split('.')[0] for s in topic_fnames]
#print(trans_ids)
```

```python
segmented_files = {}
# Key = file_id
# Value = list of lists, where each sublist is a topic
for i in topic_files:
    file_chunks = []
    topic_durations = get_topic_times(i)
    trans_id = get_trans_id(i)
    # print(trans_id)
    # print(TRANS_PATH + trans_id + '.mrt')
    trans_file = glob.glob(TRANS_PATH + trans_id + '.mrt')
    # print(trans_file)
    for t in topic_durations:
        topic_chunks = get_topic_text(t, trans_file[0])
        file_chunks.append(topic_chunks)
    segmented_files[trans_id] = file_chunks
    #print(topic_chunks)
#print(segmented_files)
```

```python
data = []
for f in segmented_files:
    for chunk in segmented_files[f]:
        #print(chunk)
        for s in range(len(chunk)):
            if chunk[s] == '':
                continue
            sent = {}
            sent['genre'] = 'transcript'
            sent['document_id'] = 'transcript_' + f
            sent['sent'] = chunk[s]
            if s == len(chunk) - 1:
                sent['boundary'] = 'yes'
            else:
                if s == len(chunk) - 2 and chunk[s+1] == '':
                    sent['boundary'] = 'yes'
                else:
                    sent['boundary'] = 'no'
            data.append(sent)
print(data[0])
```
